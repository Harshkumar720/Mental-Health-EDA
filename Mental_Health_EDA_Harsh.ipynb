## 3. ***Data Wrangling***

### Data Wrangling Code

In this step, the dataset is cleaned by handling missing values, removing irrelevant columns, and correcting inconsistent data to make it suitable for analysis.

````python
# Drop unnecessary columns safely

df.drop(['Timestamp', 'comments', 'state'], axis=1, inplace=True, errors='ignore')
````

````python
# Remove unrealistic age values

df = df[(df['Age'] >= 15) & (df['Age'] <= 80)]
df = df.copy()
````

````python
# Clean gender values properly (robust version)

df['Gender'] = df['Gender'].astype(str).str.lower().str.strip()

def clean_gender(x):
    if x.startswith('f'):
        return 'Female'
    elif x.startswith('m'):
        return 'Male'
    else:
        return 'Other'

df['Gender'] = df['Gender'].apply(clean_gender)
````

````python
df['Gender'].unique()
````

````python
# Fill missing values

df.fillna('Unknown', inplace=True)
````

````python
# Check cleaned data

df.head()
df.info()
````

The dataset contains a mix of categorical and numerical variables. Some columns have limited unique values such as treatment and family_history, indicating categorical data. Other columns like Age have a wide range of values. This step helps in understanding data diversity and identifying columns that require cleaning or transformation.

### What all manipulations have you done and insights you found?

Data wrangling was performed to clean and prepare the dataset for analysis. Unnecessary columns such as Timestamp, comments, and state were removed as they did not contribute meaningful insights. The Age column was cleaned by removing unrealistic values outside a valid range to eliminate outliers.

The Gender column was standardized by converting various inconsistent entries into three categories: Male, Female, and Other. This helped in reducing noise and improving data consistency.

Missing values in the dataset were handled by replacing them with appropriate values to ensure no null entries remained. These steps improved the overall quality of the dataset and made it suitable for further exploratory data analysis and visualization.

#### Chart - 14 - Correlation Heatmap
